## Exercise: stimulating the decision making circuit

The subpopulation 'left' and 'right' receive input form two different Poisson processes, denoted as `PoissonGroups`. They are controlled by the paremter $c$, which stands for the coherence level. For every 30 ms, the firing rate of the Poisson inputs $v_\text{left}$ and $\nu_\text{right}$ are drawn from a normal distribution: 

\begin{align*}
\nu_\text{left} \sim \mathcal{N}(\mu_\text{left}, \sigma^2) \\ 
\nu_\text{rigth} \sim \mathcal{N}(\mu_\text{right}, \sigma^2) \\
\mu_\text{left} = \mu_0  \times (0.5 + 0.5c) \\
\mu_\text{right} = \mu_0 \times (0.5 - 0.5c) \\
c \in [-1, +1]
\end{align*}



**Q1: From the equation above, express the difference $\mu_\text{left} - \mu_\text{right}$ in terms of $\mu_0$ and $c$**

$$
\mu_\text{left} - \mu_\text{right} = 1c
$$

**Q2: Find the distribution of the difference $\nu_\text{left} - \nu_\text{right}$**

Rules for adding or subtracting normally distributed variables:

- The difference between two Gaussian is another Gaussian. 
 - The means are subtracted from each other
 - The variance is added (no matter whether the normally distribtued variables are added or subtracted from each other)
 
$$
\nu_\text{left} - \nu_\text{right} \sim \mathcal{N}(\mu_\text{left} - \mu_\text{right}, 2 \sigma^2) 
$$
