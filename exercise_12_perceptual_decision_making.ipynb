## Exercise: stimulating the decision making circuit

The subpopulation 'left' and 'right' receive input form two different Poisson processes, denoted as `PoissonGroups`. They are controlled by the paremter $c$, which stands for the coherence level. For every 30 ms, the firing rate of the Poisson inputs $v_\text{left}$ and $\nu_\text{right}$ are drawn from a normal distribution: 

\begin{align*}
\nu_\text{left} \sim \mathcal{N}(\mu_\text{left}, \sigma^2) \\ 
\nu_\text{rigth} \sim \mathcal{N}(\mu_\text{right}, \sigma^2) \\
\mu_\text{left} = \mu_0  \times (0.5 + 0.5c) \\
\mu_\text{right} = \mu_0 \times (0.5 - 0.5c) \\
c \in [-1, +1]
\end{align*}



**Q1: From the equation above, express the difference $\mu_\text{left} - \mu_\text{right}$ in terms of $\mu_0$ and $c$**

$$
\mu_\text{left} - \mu_\text{right} = 1c
$$

**Q2: Find the distribution of the difference $\nu_\text{left} - \nu_\text{right}$**

Rules for adding or subtracting normally distributed variables:

- The difference between two Gaussian is another Gaussian. 
 - The means are subtracted from each other
 - The variance is added (no matter whether the normally distribtued variables are added or subtracted from each other)
 
$$
\nu_\text{left} - \nu_\text{right} \sim \mathcal{N}(\mu_\text{left} - \mu_\text{right}, 2 \sigma^2) 
$$
 
 
 

**Q3: Look at the documentation of the function `sim_decision_making_network()` and find the default values of $\mu_0$ and $\sigma$**

- Default value of $\mu_0$ : 160 (Hz), defined by the variable `mu0_mean_stimulus_Hz`
- Default value of $\sigma$ : 20 (Hz), defined by the variable `stimulus_std_Hz`

**Q4: What are the mean firing rates (in Hz) $\mu_\text{left}$ and $\mu_\text{right}$ for the coherence level $c = -0.2$?**

\begin{align*}
\mu_\text{left} = 160 \times (0.5 + 0.5 \times -0.2) = 64 \\
\mu_\text{right} = 160 \times (0.5 - 0.5 \times -0.2) = 96
\end{align*}

**Q5: For $c=0.2$, how does the difference $\mu_\text{left} - \mu_\text{right}$ compare to the variance of $\nu_\text{left} - \nu_\text{right}$?**

The difference is $64 - 96 = -32$, and the variance is $20 + 20 = 40$. The variance is slightly larger, but overall the magnitude are quite similar.

**Q7: Using your insights from the previous questions, implement a function get_decision_time that takes two RateMonitors , a avg_window_width and a rate_threshold. The function should return a tuple (decision_time_left, decision_time_right). The decision time is the time index when some decision boundary is crossed. Possible return values are (1234.5ms, 0ms) for decision Left, (0ms, 987.6ms) for decision Right and (0ms, 0ms) for the case when no decision is made within the simulation time. A return value like (123ms, 456ms) is an error and occurs if your function is called with inappropriate values for avg_window_width and rate_threshold.**

```python
# testing how to access the decision time properly 
rate_monitor_A = results["rate_monitor_A"]
avg_window_width=100 * b2.ms
rate_threshold=10 * b2.Hz
pop_A_smooth = rate_monitor_A.smooth_rate(window="flat", width=avg_window_width)/b2.Hz
np.where(pop_A_smooth > 10)[0][0]
# I think the sampling rate is 10,000Hz (10,000 samples, 1000ms = 1 second)
plt.figure()
plt.plot(pop_A_smooth)
plt.show()
```

```python
def get_decision_time(rate_monitor_A, rate_monitor_B, avg_window_width=1 * b2.ms, rate_threshold=10 * b2.Hz):
    pop_A_smooth = rate_monitor_A.smooth_rate(window="flat", width=avg_window_width)/b2.Hz
    pop_B_smooth = rate_monitor_B.smooth_rate(window="flat", width=avg_window_width)/b2.Hz
    # assume sampling rate of 10,000 Hz, ie. one sample is 0.1ms
    try:
        decision_time_left_index = np.where(pop_A_smooth > rate_threshold/b2.Hz)[0][0]
        # this exploits the fact that the above will return an error if matches found
        # we are dividing by b2.Hz just to remove the unit
        decision_time_left = decision_time_left_index / 10 * b2.ms
    except:
        decision_time_left = 0 * b2.ms
    
    try:
        decision_time_right_index = np.where(pop_B_smooth > rate_threshold/b2.Hz)[0][0]
        decision_time_right = decision_time_right_index / 10 * b2.ms
    except:
        decision_time_right = 0 * b2.ms
    
    return (decision_time_left, decision_time_right)
```

**Q3: Write a function that takes `coherence_levels, time_to_A, time_to_B, count_A, count_B, count_No` and writes `Percent correct` for each level in `coherence_levels` to the terminal**

```python
import pandas as pd
def cal_correct(coherence_levels, time_to_A, time_to_B, count_A, count_B, count_No):
    correct_choice = np.array(coherence_levels)
    coherence_levels = np.array(coherence_levels)
    
    A_correct_index = np.where(coherence_levels > 0)[0]
    correct_choice[A_correct_index] = 0 # left (A is correct)
    B_correct_index = np.where(coherence_levels < 0)[0]    
    correct_choice[B_correct_index] = 1 # right (B is correct)
        
    # probably don'tneed to use try and except
    num_trial = time_to_A.shape[1]
    
    percent_correct = list()
    
    trial = 0
    for stimuli in correct_choice:
        if stimuli == 0: # if A is correct, calculate number of time A is chosen
            percent_correct.append(count_A[trial] / float(num_trial) * 100) 
        elif stimuli == 1:
            percent_correct.append(count_B[trial] / float(num_trial) * 100)
        trial += 1
    
    percent_correct_df = {'coherence': coherence_levels, 'percent_correct': percent_correct}
    
    return pd.DataFrame(percent_correct_df)
```

```python
percent_correct_df = cal_correct(coherence_levels, time_to_A, time_to_B, count_A, count_B, count_No)
```

**Q4: Think about other values you could get from the data. Add them to your function**

You can also get the mean decision time from the trials, which may be of interest in the analysis. Note here I calculate the decision time regardless of whether the decision was correct or not. I can also do them separately, but that will be more work.

```python
def analyse_response(coherence_levels, time_to_A, time_to_B, count_A, count_B, count_No):
    correct_choice = np.array(coherence_levels)
    coherence_levels = np.array(coherence_levels)
    
    A_correct_index = np.where(coherence_levels > 0)[0]
    correct_choice[A_correct_index] = 0 # left (A is correct)
    B_correct_index = np.where(coherence_levels < 0)[0]    
    correct_choice[B_correct_index] = 1 # right (B is correct)
        
    # probably don't need to use try and except
    num_trial = time_to_A.shape[1]
    
    percent_correct = list()
    
    trial = 0
    for stimuli in correct_choice:
        if stimuli == 0: # if A is correct, calculate number of time A is chosen
            percent_correct.append(count_A[trial] / float(num_trial) * 100) 
        elif stimuli == 1:
            percent_correct.append(count_B[trial] / float(num_trial) * 100)
        trial += 1
    
    decision_time = np.concatenate((time_to_A, time_to_B), axis = 1) # join horizontally
    decision_time[decision_time == 0] = np.nan # zero are not real trials
    mean_decision_time = np.nanmean(decision_time, axis=1)
    
    response_df = {'coherence': coherence_levels, 'percent_correct': percent_correct,
                   'mean_decision_time': mean_decision_time}
    
    return pd.DataFrame(response_df)
```

```python
response_df = analyse_response(coherence_levels, time_to_A, time_to_B, count_A, count_B, count_No)
print(response_df)
```
